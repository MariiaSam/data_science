# ***2***

***Отримай середнє і стандартне відхилення віку для людей, які заробляють > 50K на рік. Після цього, отримай його для тих, хто заробляє <= 50K.***

```python
mean_more_50 = df[df["salary"].str.contains(">50K")]['age'].mean()
print("середнє відхилення віку для людей, які заробляють >50K на рік:", round(mean_more_50, 2))
print("\n")

std_more_50 = df[df["salary"].str.contains(">50K")]['age'].std()
print("стандартне відхилення віку для людей, які заробляють >50K на рік:", round(std_more_50, 2))
print("\n")

mean_less_50 = df[df["salary"].str.contains("<=50K")]['age'].mean()
print("середнє відхилення віку для людей, які заробляють <=50K на рік:", round(mean_less_50, 2))
print("\n")

std_less_50 = df[df["salary"].str.contains("<=50K")]['age'].std()
print("стандартне відхилення віку для людей, які заробляють <=50K на рік:", round(std_less_50, 2))
```
