# ***Перетворення даних***

```python
df["Order_Date"] = df["Order_Date"] + " " +  df["Time"]
df
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"], format="%Y-%m-%d %H:%M:%S")
```

```python
df.info()
```

```python
df = df.drop("Time", axis=1)
df.head()
```

```python
df["sales_amount"] = df["Sales"] * df["Quantity"]
df.head()
```

```python
df["Gender"] = df["Gender"].replace({"Male": "M", "Female": "F"})
```

```python
df["Gender"].unique()
```

```python
df.columns = df.columns.str.lower()
df.head()
```

```python
df["product_category"].unique()
```

```python
df["product_category"].nunique()
```

```python
category_stats = df.groupby("product_category").agg(
    orders_count=("product_category", "count"),
    total_quantity=("quantity", "sum"),
    total_sales=("sales_amount", "sum")
).reset_index()

print(category_stats)
```

* Скільки категорій продукції продає компанія – 4
