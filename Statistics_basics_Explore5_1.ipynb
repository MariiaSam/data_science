```python
import os
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mu, sigma = 1, 0.1
s = np.random.normal(mu, sigma, 1000)

prob, bins, ignored = plt.hist(s, 50, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(bins - mu)** 2 / (2 * sigma**2) ),linewidth=2, color='r')

plt.show()
```

```python
 # random variable, binominal distribution
 # x = number of times we throw heads in 10 coin flips
 # y = probability to see x heads

 n, p = 10, 0.8 # number of flips, probability of each result in a flip
 s = np.random.binomial(n, p, 1000000) # 1000 trials

 prob, bins, ignored = plt. hist(s, n, density=True)
 plt.show()
```

```python
# Random var, normal distribution - in more detail

# x = meat consumption from Silpo over the last 1000 days
# y = probability that x meat amount will be bougth on a given day

mu, sigma = 1, 0.1
x = np.random.normal(mu, sigma, 1000)

prob, bins, ignored = plt.hist(x, 50, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(bins - mu)** 2 / (2 * sigma**2) ),linewidth=2, color='r')

plt.show()
```

```python
from scipy.stats import norm

probability_pdf = norm.pdf(0.85, loc=mu, scale=sigma)
print(probability_pdf)
```

```python
# average mean

x.mean()
```

```python
# variance

np.sum(((x - x.mean())**2)/len(x))
```

```python
x
```

```python
# standart deviation (відхилення)
np.sqrt(np.sum(((x - x.mean())**2)/len(x)))
```

```python
# standart deviation
np.std(x)
```

```python
# new var: z = number of active coupons for discount on meat
# Poisson distribution is a discrete distribition that counts the frequency of occurrences as integers, whose list (0, 1, 2, ...)

z = np.random.poisson(500, 1000)
prob, bins, ignored = plt.hist(z, 50, density=True)
plt.show()
```

```python
# Central limit theorem

N =100000
n = 100
s = np.random.binomial(10, 0.8, N)
s = s.reshape((N//n, n))

prob, bins, ignored = plt.hist(s[0], 10, density=True)
plt.show()
```

```python
len(s)
```

```python
s_means = s.mean(axis=1)
```

```python
sigma = s_means.std()
mu = s_means.mean()

prob, bins, ignored = plt.hist(s_means, 100, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2* np.pi)) * np.exp(-(bins-mu)**2/(2*sigma**2)),
         linewidth=2, color='r')
plt.show()
```
