Ти маєш результати A/B-тесту для інтернет-магазину. Вибірка містить інформацію про відсоток конверсій для двох груп:

клієнти, які використовують стару програму лояльності (група A);
клієнти, які використовують нову програму лояльності (група B).
Твоє завдання — проаналізувати конверсії та порівняти ефективність нової програми лояльності із попередньою.



```python
from scipy.stats import mannwhitneyu

conversion_rate_A = [5.2, 4.8, 6.1, 5.0, 5.3, 4.9, 5.1, 5.4, 4.7, 5.0]
conversion_rate_B = [3, 3, 2, 4, 4]

u_stat, p_value = mannwhitneyu(conversion_rate_A, conversion_rate_B)
print(f"U-statistic: {u_stat}, p-value: {p_value}")
```

p-value: 0.002 – значно менше за 0.05.Нова програма лояльності має краще показники конверсії

У процесі тестування нової програми лояльності також збиралися дані й про рівень задоволеності клієнтів. Маємо дві групи:

група А - рівень задоволеності по шкалі від 1 до 5 для старої програми лояльності;
група В - для нової програми лояльності. Порівняй програми лояльності за цим критерієм. Чи є нова програма ефективнішою?

```python
from scipy.stats import mannwhitneyu

satisfaction_A = [3, 4, 5, 3, 4, 3, 4, 5, 3, 4]
satisfaction_B = [4, 5, 5, 4, 5, 5, 4, 5, 5, 4]

u_stat, p_value = mannwhitneyu(satisfaction_A, satisfaction_B)
print(f"U-statistic: {u_stat}, p-value: {p_value}")
```

p-value: 0.02 – значно менше за 0.05.Рівень задоволеності клієнтів також значно вищий у нвцій програмі
