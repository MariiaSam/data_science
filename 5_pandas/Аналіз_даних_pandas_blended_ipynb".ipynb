#cap-diameter (діаметр шапинки):
 зазвичай вимірюється в сантиметрах або міліметрах. Діаметр шапинки є важливою характеристикою для визначення виду гриба.
#cap-shape (форма шапинки):
Описує форму шапинки гриба. Види форм можуть включати плоску (flat), опуклу (convex), дзвоноподібну (bell-shaped), конічну (conical) тощо. Форма шапинки також допомагає в ідентифікації гриба.
#gill-attachment (кріплення пластинок):

спосіб, яким пластинки під шапинкою гриба прикріплені до ніжки. Види кріплення можуть бути вільними (free), прирослими (attached), спадними (decurrent) тощо. Ця характеристика важлива для класифікації грибів.
#gill-color (колір пластинок):
може варіюватися від білого до чорного, включаючи відтінки жовтого, коричневого, рожевого і т.д. Колір пластинок може змінюватися з віком гриба, і це також враховується при ідентифікації.
#stem-height (висота ніжки):
довжина ніжки гриба від основи до шапинки, зазвичай вимірюється в сантиметрах або міліметрах. Висота ніжки допомагає розрізняти різні види грибів.
#stem-width (ширина ніжки):
 діаметр ніжки гриба, який також вимірюється в сантиметрах або міліметрах. Ширина ніжки, як і висота, може варіюватися в залежності від виду гриба.
#stem-color (колір ніжки):
може бути білим, коричневим, жовтим, червонуватим і т.д. Колір ніжки може бути важливим критерієм для ідентифікації виду.
#season (сезон):
Час року, коли гриб з'являється або росте. Наприклад, весна, літо, осінь або зима. Багато видів грибів мають певний сезон росту, що допомагає в їх визначенні.
#class (клас):
категорія, яка може використовуватися для класифікації грибів за їх токсичністю або їстівністю. Наприклад, гриби можуть класифікуватися як їстівні, неїстівні, отруйні тощо.

Знайти гриб із мін розміром шапинки

````python
min_cap = df_sorted_by_cap.tail(1)
min_cap
````

Знайти гриб із середнім розміром шапинки

````python
# Для цього визначаємо середній розмір і знаходимо найближчий до цього значення гриб
mean_cap_value = df['cap-diameter'].mean()
mean_cap = df.iloc[(df['cap-diameter'] - mean_cap_value).abs().argsort()[:1]]
mean_cap
````


**StudentID** : унікальний ідентифікатор, призначений кожному студенту (від 1001 до 3392).
Демографічні відомості
**Вік**: Вік учнів від 15 до 18 років.
**Стать** : стать студентів, де 0 означає чоловічу, а 1 жіночу.
**Етнічна приналежність** : Етнічна приналежність студентів, закодована таким чином:
0: кавказька
1: афроамериканець
2: азіатський
3: Інше
**ParentalEducation** : рівень освіти батьків, закодований таким чином:
0: Жодного
1: Середня школа
2: Якийсь коледж
3: Бакалавр
4: вище
**Звички до навчання**
**StudyTimeWeekly** : щотижневий час навчання в годинах від 0 до 20.
**Прогули**: кількість пропусків протягом навчального року від 0 до 30.
**Репетиторство** : статус репетиторства, де 0 означає Ні, а 1 означає Так.
Участь батьків
**ParentalSupport** : рівень батьківської підтримки, закодований таким чином:
0: Жодного
1: Низький
2: Помірний
3: Високий
4: Дуже високий
**Позакласна діяльність**
Позакласний : Участь у позашкільних заходах, де 0 означає Ні, а 1 означає Так.
Спорт : участь у спортивних змаганнях, де 0 означає ні, а 1 означає так.
Музика : участь у музичних заходах, де 0 означає ні, а 1 означає так.
Волонтерська діяльність : участь у волонтерській діяльності, де 0 означає ні, а 1 означає так.
Академічна успішність
**GPA** : середній бал за шкалою від 2,0 до 4,0, на який впливають звички до навчання, участь батьків і позакласна діяльність.
Цільова змінна: оцінка класу
**GradeClass** : Класифікація оцінок студентів на основі GPA:
0: "A" (середній бал >= 3,5)
1: «B» (3,0 <= GPA < 3,5)
2: «C» (2,5 <= GPA < 3,0)
3: «D» (2,0 <= GPA < 2,5)
4: «F» (середній бал < 2,0)

````python
import pandas as pd
import requests

url = 'https://uk.wikipedia.org/wiki/Економіка_України'
df = pd.read_html(requests.get(url).text, encoding='utf-8')[11]
df
````

````python
df = df.drop(columns=[col for col in df.columns if col.startswith('Unnamed') and col != 'Unnamed: 0'])
df
````

````python
columns = df.columns

new_columns = [f"{col}, %" for col in columns]

# Перейменувати стовпці
df.columns = new_columns
df
````

````python
df.rename(columns={"Країна, %": "Країна"}, inplace=True)
df
````
