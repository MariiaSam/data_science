```python
import matplotlib.pyplot as plt
```

```python
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
visiters = [120, 150, 130, 170, 200, 230, 180]

plt.plot(days, visiters, marker="o")
plt.title("The number of visitors by day of the week")
plt.xlabel("Day of The Week")
plt.ylabel("Visiters")

plt.show()
```
