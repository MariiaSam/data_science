You manage an online store and are testing three different ad display variants for customers (A, B, C). The goal is to determine which one is most effective by the number of ad clicks (CTR — Click-Through Rate).

Perform an ANOVA test on the three different versions of the advertising campaign. Determine whether there is a significant difference between the campaigns.

**Data Description:**

* `Variant`: Ad version (A, B, or C).
* `Impressions`: Number of ad impressions (how many times the ad variant was shown to users).
* `Clicks`: Number of ad clicks (how many users clicked the ad after seeing it).
* `CTR (Click-Through Rate)`: The ratio of clicks to impressions, expressed as a percentage.


```python
import pandas as pd
from scipy.stats import f_oneway
```

```python
data = {
    "Variant": ["A", "A", "A", "B", "B", "B", "C", "C", "C", "A", "B", "C", "A", "B", "C"],
    "Impressions": [100, 150, 120, 90, 130, 140, 110, 100, 130, 160, 180, 120, 140, 110, 100],
    "Clicks": [10, 12, 14, 9, 15, 16, 11, 10, 14, 13, 17, 12, 15, 14, 13]
}
df = pd.DataFrame(data)

df["CTR (%)"] = (df["Clicks"] / df["Impressions"]) * 100

print(df)
```

```python
ctr_A = df[df["Variant"] == "A"]["CTR (%)"]
ctr_B = df[df["Variant"] == "B"]["CTR (%)"]
ctr_C = df[df["Variant"] == "C"]["CTR (%)"]

f_stat, p_value = f_oneway(ctr_A, ctr_B, ctr_C)

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")
```

p-value > 0.05, то немає достатніх підстав стверджувати, що кампанії відрізняються за CTR
