```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

# ***2.***

***Візуалізуй взаємозв'язок між розміром чайових tip та розміром чека total_bill за допомогою діаграми розсіювання.***

```python
plt.scatter(tips["total_bill"], tips["tip"], color="blue")
plt.title("Tip vs total_bill")
plt.xlabel("Total_bill")
plt.ylabel("Tip ($)")
plt.show()
```

# ***3***

***Розрахуй коефіцієнти кореляції Пірсона та Спірмена між розміром чайових та розміром чека. Чи є взаємозв'язок між ними статистично значущим?***


```python
print(f"\nTest statistic and p-value pearsonr: {pearsonr(tips['total_bill'], tips['tip'])}")

pearson_corr = tips["total_bill"].corr(tips["tip"])
print(f"\nPearson Correlation pearsonr: {pearson_corr}")

print(f"\nTest statistic and p-value spearmanr: {spearmanr(tips['total_bill'], tips['tip'])}")

spearman_corr = tips["total_bill"].corr(tips["tip"], method="spearman")
print(f"\nSpearman Correlation spearmanr: {spearman_corr}")
```

Коефіцієнт Пірсона:

r - 0.675 -сильний позитивний зв'язок

pvalue 6.69е-34 - дуже малий,  зв'язок статистино значущий



---
Коефіцієнт Спірмена:

r - 0.679 -сильний позитивний зв'язок

pvalue 2.50е-34 - дуже малий,  зв'язок статистино значущий.



---

обидва коефіцієнти мають майже однаковий результат, залежність: позитивна, сильна, статистично значуща


# ***4***

***За допомогою візуалізації визнач, чи схожий розподіл змінних tip та total_bill на нормальний. Який з двох коефіцієнтів кореляції більше підходить під ці вхідні дані?***



```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(tips["total_bill"], color="skyblue", edgecolor="black")
plt.title("Розподіл total_bill")
plt.xlabel("total_bill ($)")
plt.ylabel("Count")

plt.subplot(1, 2, 2)
plt.hist(tips["tip"], color="salmon", edgecolor="black")
plt.title("Розподіл tip")
plt.xlabel("Tip ($)")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

total_bill - правостороння асиметрія,

tip - теж має асиметрію - правий хвіст.

Розподіли не є симетричними. Отже, коефіцієнт Спірмена доречно застосовувати до змнінних total_bill та tip.

# ***5***

***Розрахуй ступінь тісноти статистичного зв'язку між розміром групи size та розміром чека total_bill. Визнач самостійно, який із коефіцієнтів кореляції тут доречно застосувати. Чи є зв'язок статистично значущим?***



```python
print(f"\nTest statistic and p-value pearsonr: {pearsonr(tips['size'], tips['total_bill'])}")

pearson_corr = tips["size"].corr(tips["total_bill"])
print(f"\nPearson Correlation pearsonr: {pearson_corr}")


print(f"\nTest statistic and p-value spearmanr: {spearmanr(tips['size'], tips['total_bill'])}")

spearman_corr = tips["size"].corr(tips["total_bill"], method="spearman")
print(f"\nSpearman Correlation spearmanr: {spearman_corr}")
```

Коефіцієнт Пірсона:

r - 0.598 -  середньо-сильний позитивний зв’язок

pvalue - 4.39e-25 - дуже малий, статистично значущий


---
Коефіцієнт Спірмена:

r - 0.605 -  середньо-сильний позитивний зв’язок
pvalue - 9.92e-26 - дуже малий,  зв'язок статистино значущий.



---
Враховуючи те, що зміна size - дискретна, то більш доречно використовувати коефіцієнт Спірмена.
