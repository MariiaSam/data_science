```python
import pandas as pd
```

```python
book_data = {
    "book_id": [201, 202, 203, 204, 205],
    "title": ["Python for Beginners", "Data Science with Python", "Advanced Python Programming", "Learning SQL", "Python for Data Analysis"],
    "author": ["John Doe", "Jane Smith", "John Doe", "Alice Johnson", "Jane Smith"],
    "price": [250, 300, 150, 180, 280],
    "copies_sold": [150, 200, 100, 90, 300]
}

df = pd.DataFrame(book_data)
print(df)
```

# ***1.***

***Книги, ціна яких більше ніж 200.***

```python
books_more_200 = df[df["price"] > 200]
books_more_200
```

# ***2.***


***Книги, ціна яких не перевищує 200, а кількість проданих копій більше або дорівнює 100. Окремо знайди кількість таких книг.***

```python
books_more_200_copy_100 = df[(df["price"] <= 200) & (df["copies_sold"] >= 100)]
books_more_200_copy_100
```

```python
print("кількість таких книг:", len(books_more_200_copy_100))
```

# ***3.***


***Книги, написані John Doe та Jane Smith.***

```python
books_author = df[df["author"].isin(["ohn Doe", "Jane Smith"])]
books_author
```

# ***4.***

***Книги, у назві яких є слово "Python".***


```python
books_python = df[df["title"].str.contains("Python")]
books_python
```
