За допомогою тестів Д'Агостіно-Пірсона та Шапіро-Уілка перевір нормальність розподілу обох вибірок.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import normaltest, shapiro

salaries = [3000, 3200, 3500, 3700, 3100, 4000, 3300, 4500, 2900, 4200, 3400, 3100,
            2950, 3600, 3800, 5000, 3700, 5200, 4700, 5500, 4300, 6000, 3100, 2800,
            5000, 3200, 4100, 3900, 4500, 3000]
work_experience = [5.9, 4.9, 5.6, 7.3, 4.8, 4.8, 7.7, 6.6, 3.2, 6.3, 3.9, 3.9, 5.5,
                   3.9, 7.3, 6.7, 5.3, 2.9, 6.6, 5.1, 4.6, 4.6, 5.6, 8.5, 3.7, 4.4,
                   6.4, 4.5, 5.5, 3.6]
```

```python
_, p_salaries_normaltest = normaltest(salaries)
_, p_exp_normaltest = normaltest(work_experience)

_, p_salaries_shapiro = shapiro(salaries)
_, p_exp_shapiro = shapiro(work_experience)

print(f"Normaltest для ЗП p-value: {p_salaries_normaltest}")
print(f"Normaltest для досвіду роботи p-value: {p_exp_normaltest}")

print(f"Shapiro-Wilkдля ЗП p-value: {p_salaries_shapiro}")
print(f"Shapiro-Wilk  для досвіду роботи data p-value: {p_exp_shapiro}")

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(salaries, bins=30, color="skyblue", edgecolor="black")
plt.title("Розподіл зарплат IT-спеціалістів")
plt.xlabel("Зарплата ($)")
plt.ylabel("Кількість")

plt.subplot(1, 2, 2)
plt.hist(work_experience, bins=30, color="salmon", edgecolor="black")
plt.title("Розподіл зарплат IT-спеціалістів")
plt.xlabel("Зарплата ($)")
plt.ylabel("Кількість")

plt.tight_layout()
plt.show()
```

ЗП. Normaltest p-value 0.18 - немає підстав відхиляти нормальність

Shapiro-Wilk p-value 0.03 - дані не відповідають нормальному розподілу.

Досвід роботи: обидва тести показують, що розподіл наближений до нормального.
