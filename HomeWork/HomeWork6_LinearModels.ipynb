```python
def hypothesis(X, w):
    return np.dot(X, w)

# X â€” Ñ†Ðµ Ð¼Ð°Ñ‚Ñ€Ð¸Ñ†Ñ Ð²Ñ…Ñ–Ð´Ð½Ð¸Ñ… Ð´Ð°Ð½Ð¸Ñ…
# w â€” Ð²ÐµÐºÑ‚Ð¾Ñ€ Ð²Ð°Ð³
```

# ***2. Ð¤ÑƒÐ½ÐºÑ†Ñ–Ñ Ð´Ð»Ñ Ð¾Ð±Ñ‡Ð¸ÑÐ»ÐµÐ½Ð½Ñ Ñ„ÑƒÐ½ÐºÑ†Ñ–Ñ— Ð²Ñ‚Ñ€Ð°Ñ‚ Ñƒ Ð²ÐµÐºÑ‚Ð¾Ñ€Ð½Ð¾Ð¼Ñƒ Ð²Ð¸Ð³Ð»ÑÐ´Ñ–***

**(Mean Squared Error, MSE)**

```python
def compute_cost(X, y, w):
    m = len(y)  # ÐºÑ–Ð»ÑŒÐºÑ–ÑÑ‚ÑŒ Ð¿Ñ€Ð¸ÐºÐ»Ð°Ð´Ñ–Ð²
    error = hypothesis(X, w) - y
    cost = (1 / (2 * m)) * np.dot(error.T, error)
    return cost
```

# ***3. ÐžÐ´Ð¸Ð½ ÐºÑ€Ð¾Ðº Ð³Ñ€Ð°Ð´Ñ–Ñ”Ð½Ñ‚Ð½Ð¾Ð³Ð¾ ÑÐ¿ÑƒÑÐºÑƒ***

```python
def gradient_step(X, y, w, alpha):
    m = len(y)
    error = hypothesis(X, w) - y
    gradient = (1 / m) * np.dot(X.T, error)
    w = w - alpha * gradient
    return w
```

```python
def gradient_descent(X, y, w, alpha, num_iters):
    cost_history = []

    for i in range(num_iters):
        w = gradient_step(X, y, w, alpha)
        cost = compute_cost(X, y, w)
        cost_history.append(cost)

    return w, cost_history
```

```python
def analytical_solution(X, y):
    X_transpose = np.transpose(X)
    w = np.linalg.inv(X_transpose.dot(X)).dot(X_transpose).dot(y)
    return w
```
