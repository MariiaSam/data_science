```python
from scipy import stats

test_group = [1, 2, 1, 3, 3, 2, 1]
loyalty_group = [3, 4, 3, 5, 4, 4]

t_stat, p_value = stats.ttest_ind(test_group, loyalty_group)

print(f"p-value: {p_value}")
```

Висновок: враховючи те, що p-value = 0.0014 < 0.05, нова програма лояльності має статистично значущий вплив на повторні покупки
