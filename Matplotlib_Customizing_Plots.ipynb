```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

df.head()
```

# ***1***

***Використовуючи набір даних про чайові, створи графік, що показує середній розмір чайових для кожного дня тижня. Подумай, який графік тут найкраще підходить.***

```python
mean_tip_by_day = df.groupby("day")["tip"].mean().sort_values(ascending=False)
mean_tip_by_day
```

```python
mean_tip_by_day.plot(kind="line", marker="o")
plt.title("Mean Tip Amount by Day of the Week")
plt.xlabel("Day of The Week")
plt.ylabel("Tip Amount ($)")

plt.show()
```

```python
mean_tip_by_day.plot(kind="bar")
plt.title("Mean Tip Amount by Day of the Week" )
plt.xlabel("Day of the Week")
plt.ylabel("Tip Amount ($)")
plt.xticks(color="grey")
plt.yticks(color="grey")

plt.show()
```

```python
mean_tip_by_day.plot(kind="barh")

plt.title("Average Tip by Day (Horizontal Bar)")
plt.xlabel("Average Tip ($)")
plt.ylabel("Day of the Week")
plt.tight_layout()
plt.show()
```

# ***3.***

***Зміни колір та розмір графіка, щоб він був більшим та зручнішим для перегляду.***



```python
plt.figure(figsize=(10, 6))

mean_tip_by_day.plot(kind="line", marker="o", color="darkorange")
plt.title("Mean Tip Amount by Day of the Week", fontsize=16, fontweight='semibold', color='blue', fontstyle= "italic", ha="center")
plt.xlabel("Day of The Week", fontweight='light', color='blue', fontstyle= "italic")
plt.ylabel("Tip Amount ($)", fontweight='light', color='blue', fontstyle= "italic")
plt.xticks(color="grey",fontweight='bold', fontstyle='normal', fontsize=10)
plt.yticks(color="grey",fontweight='bold', fontstyle='normal', fontsize=10)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

mean_tip_by_day.plot(kind="bar", color="coral")
plt.title("Mean Tip Amount by Day of the Week", fontsize=16, fontweight='semibold', color='blue', fontstyle= "italic", ha="center" )
plt.xlabel("Day of the Week", fontweight='light', color='blue', fontstyle= "italic")
plt.ylabel("Tip Amount ($)", fontweight='light', color='blue', fontstyle= "italic")
plt.xticks(color="grey",fontweight='bold', fontstyle='normal', fontsize=10)
plt.yticks(color="grey",fontweight='bold', fontstyle='normal', fontsize=10)

plt.show()
```
