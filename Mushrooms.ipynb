Сортуємо гриби за розміром шапинки (cap-diameter) у порядку спадання

# Вибираємо три гриби з найбільшим розміром шапинки

```python
df_cap_shape_3 = df_cap_diameter.head(3)
df_cap_shape_3
```

Знайти гриб із мін розміром шапинки

```python
min_cap_shape = df_cap_diameter.tail(1)
min_cap_shape
```

Знайти гриб із середнім розміром шапинки

```python
mean_cap = df['cap-diameter'].mean()
mean_cap
```

```python
df_mean_cap = df.iloc[(df['cap-diameter'] - mean_cap).abs().argsort()[:1]]
df_mean_cap
```

Об'єднати гриби в один датасет

```python
new_df = pd.concat([df_cap_shape_3, min_cap_shape , df_mean_cap, ])
new_df
```

Видалити два останніх стовпчика

```python
new_df = new_df.iloc[:, :-2]
new_df
```
