# ***7.***

***Порівняй зарплати одружених і неодружених чоловіків. Хто заробляє більше? (> 50K або < = 50K) Одружені чоловіки — це ті, у кого marital-status починається з Married. Інші — не одружені.***

```python
m = df["marital-status"].unique()
m
```

```python
male = df["sex"] == "Male"
```

```python
married = df["marital-status"].str.startswith("Married")
```

```python
married_man = df[male & married]
unmarried_man = df[male & ~ married]

print("Одружені чоловіки -", married_man["salary"].value_counts())
print("Неодружені чоловіки -", unmarried_man["salary"].value_counts())
```

# ***8.***

***Отримай максимальну кількість годин на тиждень, яку працює людина. Скільки людей працює стільки ж годин на тиждень?***

```python
max_hours_per_week = df["hours-per-week"].max()
print("максимальна кількість годин на тиждень - ",max_hours_per_week)
```

```python
people_hours = (df["hours-per-week"] == max_hours_per_week).sum()
print("Кількіст людей, які працюють максимальну кількість годин на тиждень -", people_hours)
```
