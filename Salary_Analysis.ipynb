```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
salary = [3800, 3100, 3700, 4000, 4000, 3700, 2400, 1900, 3600, 4600,
4800, 4500, 2200, 3700, 2900, 10500, 2700, 4600, 3600, 3400,
12000, 1800, 300, 4000, 3200, 1100, 4300, 5000, 4300, 4300,
4300, 3200, 3700, 7000, 2400, 2300, 3900, 3400, 3900, 4800,
3000, 2600, 4700, 4200, 4100, 3600, 4000, 4500, 3500, 4600]
```


# ***2. Чи є викиди у цьому наборі даних? Використай відомі тобі способи виявлення викидів.***

```python
sns.lineplot(salary, marker="o")
plt.title("Dynamics of Salary")
plt.xlabel("Employee ID")
plt.ylabel("Salary ($)")
plt.show()
```

```python
Q1 = np.percentile(salary, 25)
Q3 = np.percentile(salary, 75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in salary if x < lower_bound or x > upper_bound]
print("Викиди:", outliers)
```

Викиди в значеннях 300, 1100, 7000, 10500, 12000($)

# ***3. Побудуй гістограму для виявлення розподілу заробітної плати.***

```python
plt.hist(salary, bins=10, color="blue", alpha=0.7)
plt.title("Distribution of Salary")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

# ***4. Проаналізуй гістограму. Чи схожий розподіл даних та нормальний розподіл? На рівномірний розподіл? Чи є розподіл симетричним?***

1. Нормальний розподіл має дзвонопобідну форму.
У моєму випадку, більшість значень концентруються у діапазоні 2000-5000 одиниць.

Але спостегрігаються випадки дуже великої і маленької ЗП, 300, 1100, 7000, 10500, 12000.

Через ці викиди розподіл не є нормальним.

2. Рівномірний означає, що інтервали мають приблизно однакову частоту.
Концентрація значень у середньому діапазоні.Тож, розподіл не є рівномірний.

3. Розподіл має правосторонню асиметрію, більшість значень в нижньому та середньому діапазоні.  Також, є хвіст вправо з великими значеннями.


# ***5. Спробуй видалити викиди із набору даних та знову побудувати гістограму. Чи змінився характер розподілу? На який із відомих тобі розподілів він більш схожий?***

```python
clean_salary = [i for i in salary if lower_bound <= i <= upper_bound]
print("дані без викидів", clean_salary)
```

```python
plt.hist(clean_salary, bins=10, color="blue", alpha=0.7)
plt.title("Distribution of Salary")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

Після видалення викидиів розподіл став більш концентрованим у діапазоні 18000-5000.

Очищені дані нагадують до нормального розподілу
