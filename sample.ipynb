# Простий випадковий вибір

```python
import numpy as np
import pandas as pd

# Creating a simple dataset of 1000 sales
data = {
    "sale_id": np.arange(1, 1001),
    "sale_amount": np.random.randint(50, 500, size=1000),
}
sales_data = pd.DataFrame(data)

# Simple Random Sampling: select 100 random rows
sample_data = sales_data.sample(n=100, random_state=42)

sample_data.head()
```

# Стратифікована вибірка

```python
# Adding a product category to the sales data
sales_data["Category"] = np.random.choice(
    ["Electronics", "Clothing", "Home"], size=1000
)

# Stratified sampling: selecting 20 samples from each category
stratified_sample = sales_data.groupby("Category", group_keys=False).apply(
    lambda x: x.sample(20, random_state=42)
)

stratified_sample.head()
```
