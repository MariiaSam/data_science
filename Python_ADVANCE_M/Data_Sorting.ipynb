```python
import pandas as pd

# Creating a DataFrame with employee data
employee_data = {
    "employee_id": [101, 102, 103, 104, 105],
    "name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "department": ["HR", "IT", "Finance", "Marketing", "IT"],
    "salary": [50000, 60000, 55000, 70000, 65000],
    "years_of_experience": [5, 7, 4, 10, 6]
}

df = pd.DataFrame(employee_data)
print(df)
```

# ***1.***

***Зарплата від найменшої до найбільшої.***

```python
salary_min_max = df.sort_values(by="salary")
salary_min_max
```

# ***2.***


***Дані за досвідом роботи у порядку спадання.***

```python
years_of_experience = df.sort_values(by="years_of_experience", ascending=False)
years_of_experience
```

# ***3.***


***Дані спочатку за департаментом у зростаючому порядку, а потім за зарплатою у порядку спадання.***

```python
department_salary = df.sort_values(by=["department", "salary"], ascending=[True, False])
department_salary
```
