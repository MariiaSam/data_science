```python
import pandas as pd
```

```python
products = {
    "product_id": [1, 2, 3, 4],
    "product_name": ["Laptop", "Tablet", "Smartphone", "Smartwatch"],
    "category": ["Electronics", "Electronics", "Electronics", "Wearables"],
    "price": [1200, 300, 800, 250],
}

df_products = pd.DataFrame(products)
print(df_products)
```

```python
order_details = {
    "order_id": [101, 102, 103, 104, 105, 106, 107],
    "product_id": [2, 2, 1, 2, 4, 1, 5],
    "customer_id": [201, 202, 202, 204, 203, 201, 202],
    "quantity": [1, 2, 1, 3, 2, 2, 4],
}

df_order_details = pd.DataFrame(order_details)
print(df_order_details)
```

# ***1.***

***Виконай INNER JOIN між таблицею замовлень та продуктів на основі product_id. Які продукти були замовлені? Запиши відповідь у текстовому блоці..***

```python
inner_join = pd.merge(df_products, df_order_details, on="product_id", how="inner")
print(inner_join)
```

***Laptop, Tablet, Smartphone***




# ***2.***


***Виконай LEFT JOIN між таблицею замовлень та продуктів. Які продукти не були замовлені? Запиши відповідь у текстовому блоці..***

```python
left_join = pd.merge(df_products, df_order_details, on="product_id", how="left")
print(left_join)
```

***Smartphone***

# ***3.***


***Виконай OUTER JOIN між таблицею замовлень та продуктів. Якого ID продукту не вистачає у таблиці продуктів? Запиши відповідь у текстовому блоці.***

```python
outer_join = pd.merge(df_products, df_order_details, on="product_id", how="outer")
print(outer_join)
```
