```python
import pandas as pd

# Creating a dataset
data = {
    "order_id": [201, 202, 203, 204, 205, 206, 207, 208, 209, 210],
    "product": [
        "Laptop",
        "Smartphone",
        "Laptop",
        "Tablet",
        "Smartphone",
        "Tablet",
        "Smartwatch",
        "Laptop",
        "Headphones",
        "Smartwatch",
    ],
    "category": [
        "Electronics",
        "Electronics",
        "Electronics",
        "Electronics",
        "Electronics",
        "Electronics",
        "Wearables",
        "Electronics",
        "Accessories",
        "Wearables",
    ],
    "region": [
        "North",
        "South",
        "East",
        "North",
        "West",
        "South",
        "East",
        "West",
        "North",
        "West",
    ],
    "sales": [1500, 900, 1200, 600, 750, 450, 320, 1400, 200, 380],
}

df = pd.DataFrame(data)
print(df)
```

# ***1.***

***Створи зведену таблицю, яка показує загальні продажі (sales) для кожної категорії продукту (category). Яка категорія має найбільші загальні продажі? Запиши відповідь у текстовому блоці.
.***

```python
total_sum_by_category = pd.pivot_table(df, values="sales", index="category", aggfunc="sum")
total_sum_by_category.sort_values(by="sales", ascending=False).reset_index()
```

***Категорія Electronics має найбільші загальні продажі***

# ***2.***


***Створи зведену таблицю, яка показує загальні продажі для кожного продукту (product) у кожному регіоні (region). Який продукт мав найвищі продажі в регіоні North? Запиши відповідь у текстовому блоці.***

```python
total_sum_by_product_region = pd.pivot_table(df, values="sales", index=["product", "region"], aggfunc="sum")
total_sum_by_product_region.sort_values(by="sales", ascending=False).reset_index()
```

***Laptop мав найвищі продажі в регіоні North***

# ***3.***


***Створи зведену таблицю, яка показує загальні продажі та середні продажі для кожної категорії продукту. Яка середня сума продажів для категорії Electronics? Запиши відповідь у текстовому блоці.***

```python
total_sum__mean_by_category = pd.pivot_table(df, values="sales", index="category", aggfunc=["sum", "mean"])
total_sum__mean_by_category.sort_values(by=("sum", "sales"), ascending=False).reset_index()
```

***Середня сума продажів для категорії Electronics – 971.428571***
